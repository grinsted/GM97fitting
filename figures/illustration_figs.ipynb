```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('aslak')
import settings
```

```python
xo=.2
yo=.3
def box(x0,y0,h=1,w=1):
    x=np.array([0,0,1,1,0])
    y=np.array([0,1,1,0,0])
    #plt.fill(x0+x*w,y0+y*h, lw=1, facecolor='w', edgecolor='k',capstyle='round')
    kwargs = {'linewidth':1,'clip_on':False, 'color':'k','zorder': y0}
    plt.plot(x0+x*w,y0+y*h, **kwargs)
    #top:
    plt.plot(x0+x*w+y*xo,y0+h+y*yo,**kwargs)#, lw=1, facecolor='w', edgecolor='k',capstyle='round')
    plt.plot(x0+w+x*xo,y0+y*h+x*yo,**kwargs)#, lw=1, facecolor='w', edgecolor='k',capstyle='round')

box(0,0)
box(0,2,4,1)
color='m'

plt.arrow(0.5+xo/2,2,0,-1+yo/2,lw=0.3,edgecolor='none',facecolor=color,width=0.04,length_includes_head=True,color=color,zorder=1.5)
plt.text(0.5+xo/2,1.5+yo/2,r' $\sigma_{zz}$',fontsize='small',color=color)

#plt.arrow(0.5+xo/2,2,0,-1+yo/2,lw=0.3,facecolor='k',width=0.04,length_includes_head=True)
#color='r'
plt.arrow(0,4+yo/2,-1,0,lw=0.3,edgecolor='none',facecolor=color,width=0.04,length_includes_head=True,color=color,zorder=1.5)
plt.arrow(1+xo/2,4+yo/2,1,0,lw=0.3,edgecolor='none',facecolor=color,width=0.04,length_includes_head=True,color=color,zorder=2)
plt.text(0-0.5,4+yo/2,r'$\dot\epsilon_{xx}$',fontsize='small',color=color,ha='center',va='bottom')
plt.text(1+xo/2+0.5,4+yo/2,r'$\dot\epsilon_{xx}$',fontsize='small',color=color,ha='center',va='bottom')
plt.arrow(0.5,4,-xo*2,-yo*2,lw=0.3,overhang=-.2,edgecolor='none',facecolor=color,width=0.04,length_includes_head=True,color=color,zorder=2)
plt.text(0.5-xo*2,4-yo*2,r'  $\dot\epsilon_{yy}$',fontsize='small',color=color,ha='left',va='center')

#shear
plt.arrow(1-xo/4,6-yo/4,-1,0,lw=0.3,edgecolor='none',facecolor=color,width=0.04,length_includes_head=True,color=color,zorder=2)
plt.arrow(0+xo+xo/4,6+yo+yo/4,1,0,lw=0.3,edgecolor='none',facecolor=color,width=0.04,length_includes_head=True,color=color,zorder=2)
plt.text(0.5-xo/4,6-yo/2,r'$\dot\epsilon_{xy}$',fontsize='small',color=color,ha='center',va='top')
plt.text(0.5+xo+xo/4,6+yo+yo/2,r'$\dot\epsilon_{xy}$',fontsize='small',color=color,ha='center',va='bottom')

color='r'
#plt.arrow(0,1,-.5,0,lw=0.3,edgecolor='none',facecolor=color,width=0.04,length_includes_head=True,color=color,zorder=2)
plt.text(0.5,0.5,r'$\dot\epsilon_{zz}$?',fontsize='small',color=color,ha='center',va='center')


plt.axis('equal')
plt.axis('off')
plt.savefig(settings.figure_folder+'/illustration.png')
```
