```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("aslak")
from scipy.interpolate import make_lsq_spline, BSpline
from scipy.ndimage import gaussian_filter1d


def make_knot_vector(x, k=3):
    # Full knot vector with endpoint multiplicity
    t = np.concatenate((np.repeat(x[0], k), x, np.repeat(x[-1], k)))
    return t


N = 3
rho = np.linspace(300, 900, N)

t = make_knot_vector(rho, k=3)

r = np.linspace(300, 900, 1000)
y = -np.sin(r / 400)
spl = make_lsq_spline(r, y, t, k=3)


plt.plot(r, y, "o", label="data")
plt.plot(r, spl(r), "-", label="spline")
plt.legend()
```

```python
c = np.pad(spl.c[1:-1], (1, 1), mode="edge")
c[0] = c[1] * 2 - c[2]
spl.c = c
plt.plot(r, spl(r))
```

```python
da2 = np.diff(spl(r), 2)
prior_da2 = np.median(da2)

plt.plot(np.diff(spl(r), 2))
prior_da2
```
